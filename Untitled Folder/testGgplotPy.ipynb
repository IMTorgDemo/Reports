```python
import pandas as pd

mtcars = pd.read_csv('mtcars.csv')
```

```python
from ggplot import *

ggplot(mtcars, aes('mpg', 'qsec')) + geom_point(colour='steelblue') + scale_x_continuous(breaks=[10,20,30], labels=["horrible", "ok", "awesome"])
```

```python
mtcars.head()
```

```python
p = ggplot(mtcars, aes('mpg','qsec')) + geom_point()
print(p)
```

```python
mtcars['gearCat']= mtcars['gear'].astype('category')
mtcars['gearCat'].cat.categories=["3gears","4gears","5gears"]
mtcars['amCat']= mtcars['am'].astype('category')
mtcars['amCat'].cat.categories=["Automatic","Manual"]
mtcars['cylCat']= mtcars['cyl'].astype('category')
mtcars['cylCat'].cat.categories=["4cyl","6cyl","8cyl"]
```

```python
mtcars[['gear','gearCat','am','amCat','cyl','cylCat']].head()
```

```python
ggplot(mtcars, aes(x='mpg', color='gearCat')) + geom_density()
```

```python
#qplot(hp, mpg, data=mtcars, shape=am, color=am, facets=gear~cyl, size=I(3))
ggplot(mtcars, aes(x='mpg',y='hp',color='amCat', shape='amCat')) + geom_point(alpha=1/2)+ facet_wrap('gearCat','cylCat') 
```

```python
#qplot(wt, mpg, data=mtcars, geom=c('point','smooth'), method='lm', formula=y~x, color=cyl)
ggplot(aes(x='wt',y='mpg',color='cylCat'), data=mtcars ) + stat_smooth(method='lm') + geom_point()
```

```python
#p = qplot(gear, mpg, data=mtcars, geom=c('boxplot','jitter'), fill=am)
#p + theme_bw()
ggplot(aes('gear','mpg', fill='amCat'),data=mtcars) + geom_boxplot()
```
