```r
library(ggplot2)
```

```r
mtcars = read.csv('mtcars.csv')
head(mtcars)
```

```r
mtcars$gear = factor(mtcars$gear, labels=c('3gr','4gr','5gr'))
mtcars$am = factor(mtcars$am, labels=c('Auto','Manual'))
mtcars$cyl = factor(mtcars$cyl, labels=c('4cyl','6cyl','8cyl'))
```

```r
qplot(mpg, data=mtcars, geom='density', fill=gear, alpha=I(.5))
```

```r
qplot(hp, mpg, data=mtcars, shape=am, color=am, facets=gear~cyl, size=I(3))
```

```r
qplot(wt, mpg, data=mtcars, geom=c('point','smooth'), method='lm', formula=y~x, color=cyl)
```

```r
p = qplot(gear, mpg, data=mtcars, geom=c('boxplot','jitter'), fill=am)
p + theme_bw()
```
